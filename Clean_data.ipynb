```python
# Most important continuous variables to check for outliers
truly_continuous = [
    'Age at enrollment',
    'Curricular units 1st sem (grade)',
    'Curricular units 2nd sem (grade)',
    'Unemployment rate',
    'Inflation rate',
    'GDP'
]
```

```python
df_continuous = df_integrated[truly_continuous]
Q1 = df_continuous.quantile(0.25)
Q3 = df_continuous.quantile(0.75)
IQR = Q3 - Q1
outliers = ((df_continuous<(Q1-1.5*IQR))| (df_continuous>(Q3+1.5*IQR))).sum()
outliers = outliers[outliers>0].sort_values(ascending=False)
outliers
```

```python
mask = ~((df_continuous < (Q1 - 1.5 * IQR)) | 
         (df_continuous > (Q3 + 1.5 * IQR))).any(axis=1)

df_clean = df_integrated[mask].copy()
```

```python
print(f"\nOriginal: {len(df_integrated)} rows")
print(f"Clean: {len(df_clean)} rows")
print(f"Removed: {len(df_integrated) - len(df_clean)} rows ({100*(1-len(df_clean)/len(df_integrated)):.1f}%)")
```

```python
from sklearn.preprocessing import StandardScaler

# 1. Continuous variables → StandardScaler
scaler = StandardScaler()
df_clean[truly_continuous] = scaler.fit_transform(df_clean[truly_continuous])
```
